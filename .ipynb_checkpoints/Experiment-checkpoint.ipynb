### Algorithm

```python
#python
import numpy as np
import pandas as pd

data_log = []  # This will store all the data logs
iteration_count = 0  # Global variable to count the iterations

def collect_data(iteration, sensor_inputs, outputs, vl, vr, weights):
    data_log.append({
        "iteration": iteration,
        "sensor_inputs": sensor_inputs.tolist(),
        "outputs": outputs.tolist(),
        "vl": vl,
        "vr": vr,
        "weights": [w.tolist() for w in weights]  # Assuming weights is a numpy array
    })
    
def export_data():
    # Define the file path
    file_path = r'C:\Users\odans\Documents\Retrieve\simulation_data.csv'
    
    try:
        df = pd.DataFrame(data_log)
        # Append mode, write header only if the file does not exist
        df.to_csv(file_path, mode='a', header=not os.path.exists(file_path), index=False)
        data_log.clear()  # Clear the list after saving to free up memory
    except Exception as e:
        print(f"Failed to export data: {e}")        

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class MultiLayerCTRNN:
    def __init__(self, layer_sizes, weight_range, bias_range, tau_range, dt=0.05):
        self.layer_sizes = layer_sizes
        self.layers = len(layer_sizes)
        self.weights = [np.random.uniform(low=weight_range[0], high=weight_range[1], size=(layer_sizes[i], layer_sizes[i + 1])) for i in range(self.layers - 1)]
        self.biases = [np.random.uniform(low=bias_range[0], high=bias_range[1], size=layer_sizes[i]) for i in range(self.layers)]
        self.taus = [np.random.uniform(low=tau_range[0], high=tau_range[1], size=layer_sizes[i]) for i in range(self.layers)]
        self.states = [np.zeros(layer_sizes[i]) for i in range(self.layers)]
        self.outputs = [np.zeros(layer_sizes[i]) for i in range(self.layers)]
        self.dt = dt

    def step(self, inputs):
        # First layer receives external inputs
        net_input = np.dot(inputs, sigmoid(self.states[0] + self.biases[0]))
        self.states[0] += self.dt * (-self.states[0] + net_input) / self.taus[0]
        self.outputs[0] = sigmoid(self.states[0] + self.biases[0])
        
        # Subsequent layers
        for i in range(1, self.layers):
            net_input = np.dot(self.outputs[i - 1], sigmoid(self.states[i] + self.biases[i]))
            self.states[i] += self.dt * (-self.states[i] + net_input) / self.taus[i]
            self.outputs[i] = sigmoid(self.states[i] + self.biases[i])
        
        return self.outputs[-1]  # Return the output of the last layer

    def apply_homeostatic_plasticity(self, HL, HU, tau_w, tau_b, time_step):
        for layer in range(self.layers):
            # Calculate plastic facilitation rho for each neuron
            rho = np.zeros(self.layer_sizes[layer])
            for i in range(self.layer_sizes[layer]):
                z = self.outputs[layer][i]
                if z < HL:
                    rho[i] = (HL - z) / HL
                elif z > HU:
                    rho[i] = (HU - z) / (1 - HU)
                # When z is within the range [HL, HU], rho[i] remains 0
            
            # Update synaptic weights for the current layer (if not output layer)
            if layer < self.layers - 1:
                for i in range(self.layer_sizes[layer]):
                    for j in range(self.layer_sizes[layer + 1]):
                        weight_change = rho[i] * np.abs(self.weights[layer][i][j]) * time_step / tau_w
                        self.weights[layer][i][j] += weight_change if self.weights[layer][i][j] > 0 else -weight_change
            
            # Update neuron biases
            for i in range(self.layer_sizes[layer]):
                bias_change = rho[i] * time_step / tau_b
                self.biases[layer][i] += bias_change

sensor_inputs = np.zeros(2)

def sysCall_init():
    sim = require('sim')

    global leftJoint, rightJoint, leftSensor, rightSensor, ctrnn

    # Get handles
    leftJoint = sim.getObject('/DynamicLeftJoint')
    rightJoint = sim.getObject('/DynamicRightJoint')
    leftSensor = sim.getObject('/LeftSensor')
    rightSensor = sim.getObject('/RightSensor')

    # Instantiate MultiLayerCTRNN
    layer_sizes = [2, 2, 2, 2]  # Input layer, two hidden layers, and output layer
    weight_range = (-10, 10)
    bias_range = (-1, 1)
    tau_range = (1, 10)
    ctrnn = MultiLayerCTRNN(layer_sizes, weight_range, bias_range, tau_range)

def sysCall_sensing():
    global sensor_inputs
    
    # Retrieve sensor handles (consider moving this to sysCall_init if they don't change)
    leftJoint = sim.getObject('/DynamicLeftJoint')
    rightJoint = sim.getObject('/DynamicRightJoint')
    
    leftSensor = sim.getObject('/LeftSensor')
    rightSensor = sim.getObject('/RightSensor')
    
    # Read sensor values
    result_left = sim.readVisionSensor(leftSensor)
    result_right = sim.readVisionSensor(rightSensor)

    # Assuming the auxiliary packet contains the data we need
    if result_left[1]:
        avg_red_left = result_left[1][11]  # Adjust the index if needed
    else:
        avg_red_left = 0  # Default value if reading failed

    if result_right[1]:
        avg_red_right = result_right[1][11]  # Adjust the index if needed
    else:
        avg_red_right = 0  # Default value if reading failed

    # Print the average red values
    print("left = " + str(avg_red_left))
    print("right = " + str(avg_red_right))
    print(result_right[1][11])

    # Normalize inputs and store them in the global variable 'sensor_inputs'
    sensor_inputs = np.array([avg_red_left, avg_red_right]) / 255.0
 
def sysCall_actuation():
    global iteration_count, sensor_inputs

    # Get outputs from MultiLayerCTRNN using sensor inputs
    outputs = ctrnn.step(sensor_inputs)
    
    # Parameters for homeostatic plasticity
    HL = 0.4  # Lower bound for homeostasis
    HU = 0.6  # Upper bound for homeostasis
    tau_w = 100  # Time constant for synaptic weight plasticity
    tau_b = 100  # Time constant for bias plasticity
    time_step = 0.05  # Simulation time step, adjust as necessary for your simulation

    # Apply homeostatic plasticity
    ctrnn.apply_homeostatic_plasticity(HL, HU, tau_w, tau_b, time_step)
    
    # Assuming outputs is an array of length corresponding to the last layer size, which should be 2
    if len(outputs) != 2:
        print("Unexpected output size from ctrnn.step()")
        return  # Add error handling or return to avoid further errors

    # Convert neural outputs to wheel speeds
    max_speed = 0.5  # Maximum wheel speed
    vl = outputs[0] * max_speed  # Use the first element for left wheel speed
    vr = outputs[1] * max_speed  # Use the second element for right wheel speed

    # Set wheel velocities
    sim.setJointTargetVelocity(leftJoint, vl)
    sim.setJointTargetVelocity(rightJoint, vr)

    # Collect and log data
    collect_data(iteration_count, sensor_inputs, outputs, vl, vr, ctrnn.weights)

    # Increment the iteration counter
    iteration_count += 1
    
    # Periodically saving data every 100 iterations
    if iteration_count % 100 == 0:
        export_data()


def sysCall_cleanup():
    # Ensuring all remaining data is saved
    if data_log:  # Check if there's any data left to save
        export_data()
    pass
```

### path following accuracy Experiment

For the next analysis, we could look at the robot's path following accuracy. To do this, we would calculate how closely the robot follows the intended path. We have the intended path represented by two lines: one vertical and one horizontal, as shown in the plot.

Here's a proposed approach to measure the path following accuracy:

- **Segment Analysis:** Break down the robot's path into segments corresponding to the intended path segments (the straight vertical and horizontal lines).
- **Deviation Calculation:**
- For the vertical line segment, calculate the horizontal deviation (difference in x) from the intended line.
- For the horizontal line segment, calculate the vertical deviation (difference in y) from the intended line.
- **Statistical Summary:** Compile these deviations into a statistical summary, possibly using descriptive statistics like mean, median, standard deviation, and maximum deviation.
- **Visual Representation:** Optionally, create a visual representation, like a scatter plot overlaid on the intended path, to show where the deviations are most significant.

```python
# To approximate the intended path, we'll assume the straight vertical line is the first part of the path
# where the x-coordinate doesn't change much, and the horizontal line is the second part where y-coordinate doesn't change much.
# We'll find the approximate x and y coordinates for the two segments.

# Approximating the x-coordinate for the vertical line segment (mean of the smallest x-values)
vertical_line_x_approx = baseline_updated_data['average_x'].nsmallest(int(len(baseline_updated_data)/2)).mean()

# Approximating the y-coordinate for the horizontal line segment (mean of the largest y-values)
horizontal_line_y_approx = baseline_updated_data['average_y'].nlargest(int(len(baseline_updated_data)/2)).mean()

# Now calculating the deviations
baseline_updated_data['deviation_vertical'] = baseline_updated_data['average_x'].apply(lambda x: abs(x - vertical_line_x_approx))
baseline_updated_data['deviation_horizontal'] = baseline_updated_data['average_y'].apply(lambda y: abs(y - horizontal_line_y_approx))

# Statistical summary of deviations for both segments
vertical_deviation_stats = baseline_updated_data['deviation_vertical'].describe()
horizontal_deviation_stats = baseline_updated_data['deviation_horizontal'].describe()

vertical_deviation_stats, horizontal_deviation_stats
```

###  **Learning Capability Test**

```python
# Placeholder value for k, the number of iterations per run
k = 500  # Replace with your actual value

# Assuming 'complex_data' is the DataFrame that includes path deviation and speed for each iteration
# Add 'run_number' based on iteration
complex_data['run_number'] = complex_data['iteration'] // k

# Calculate the mean path deviation for each run
average_path_deviation_per_run = complex_data.groupby('run_number')['path_deviation'].mean()

# Calculate the speed variance for each run
# Here we assume that 'average_v' is the column with the speed values
speed_variance_per_run = complex_data.groupby('run_number')['average_v'].var()

# Now plot the learning curves
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

# Learning curve for path deviation
plt.subplot(1, 2, 1)
plt.plot(average_path_deviation_per_run.index, average_path_deviation_per_run.values, marker='o', linestyle='-')
plt.title('Learning Curve for Path Deviation')
plt.xlabel('Run Number')
plt.ylabel('Average Path Deviation')
#plt.grid(True)

# Learning curve for speed variance
plt.subplot(1, 2, 2)
plt.plot(speed_variance_per_run.index, speed_variance_per_run.values, marker='o', linestyle='-')
plt.title('Learning Curve for Speed Variance')
plt.xlabel('Run Number')
plt.ylabel('Speed Variance')
#plt.grid(True)

plt.tight_layout()

# Saving graph as vector_image
file_path = r"C:\Users\odans\Documents\Images\image"

plt.savefig(file_path)

# Show the plot
plt.show()

# Return the path for the user to download
file_path
```

### Results

The learning curve for path deviation shows variation in average path deviation across runs. 
The learning curve for speed variance shows a negative trend, implying that any variance in speed is reducing over time. This could indicate that while the robot's ability to regulate its speed is improving, its path-following precision might not be effectively learning or there are other factors influencing its performance that might need further investigation.

For learning to be effective, especially with CTRNN and synaptic homeostasis, one would expect to see:

- A learning curve that shows gradual improvement as the number of runs increases (for path deviation).
- A reduction in the variability of both path deviation and speed as the robot learns the task (the lines should be descending over time).

In the provided plot, the speed variance does seem to be decreasing, which is a good sign if speed regulation is part of the learning objective. It's important to note that learning can sometimes be non-monotonic, especially in complex environments or with sophisticated learning mechanisms that involve exploration.
